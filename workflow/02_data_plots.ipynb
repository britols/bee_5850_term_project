```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t_trend = pd.read_csv("data/cao_index_yearly.csv",sep=";")
t_trend = t_trend.set_index('year')
t_trend.columns = ['time_min', 'time_max', 't2min_weighted', 'cao_index']

t_trend["rolling_t2min"] = t_trend["t2min_weighted"].rolling(5,center=True).mean()
t_trend["rolling_caoindex"] = t_trend["cao_index"].rolling(5,center=True).mean()
```

```python
fig, ax = plt.subplots(1,2,sharex=True,figsize=(16, 3))

ax[0].plot(t_trend['t2min_weighted'])
ax[0].scatter(t_trend.index,t_trend['t2min_weighted'],c="#1f77b4",edgecolors='black')
ax[0].set_title('DJF minimum temperature mean');
ax[0].set_ylabel('Temperature [C]');

ax[1].plot(t_trend['cao_index'],color="#FF0000")
ax[1].scatter(t_trend.index,t_trend['cao_index'],c="#FF0000",edgecolors='black')
ax[1].set_title('CAO index annual maximum');
ax[1].set_ylabel('CAO index [C]');
```
